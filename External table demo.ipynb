## Create an External table (same DDL as )

Managed, LOCATION 'abfss://container@storageaccount.dfs.core.windows.net/customers_external'

````python
%sql
CREATE TABLE customers_external (
  id INT,
  name STRING,
  city STRING
)
LOCATION 'abfss://container@storageaccount.dfs.core.windows.net/customers_external';
````

## What DESCRIBE would show, if this succeeded

````python
%sql
### In a real enterprise environment, this next command would show:

DESCRIBE TABLE EXTENDED customers_external;

Type: EXTERNAL
Location: abfss://container@storageaccount.dfs.core.windows.net/customers_external

Compare that to our Managed table earlier — Location was blank there.
Here, it's a real, visible path. That's the core difference on display.
````

## The moment that actually matters — DROP TABLE

````python
%sql
### This is the one behavior worth remembering above everything else.

DROP TABLE customers_external;

For our Managed table, this deleted the underlying files (with a 7-day
UNDROP safety net). For an External table, this ONLY removes the entry
from Unity Catalog. The Parquet files sitting in ADLS are completely
untouched — because Databricks never owned them.

If you went and checked that ADLS container right now with Azure Storage
Explorer, the files would still be sitting there, as if nothing happened.
````

## Try to query it (reference only)

````python
%sql
```sql
SELECT * FROM customers_external;
-- Fails: TABLE_OR_VIEW_NOT_FOUND
-- The catalog entry is gone. But unlike our Managed table demo,
-- there's no UNDROP TABLE here that matters the same way — the
-- data was never at risk in the first place. You'd simply
-- re-register the table pointing at the same LOCATION, and all
-- your data reappears instantly, because it was always there.
```
````
