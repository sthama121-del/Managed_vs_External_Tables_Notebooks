## Setup: Create and use the demo schema
## 

```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.demo;
```

```python
%sql
USE workspace.demo
```

## Create a Managed table
## __

```python
%sql
CREATE TABLE customers (
  id INT,
  name STRING,
  city STRING
);
```

Reveal: Confirm it's Managed, and see where it actually lives


```python
%sql
DESCRIBE TABLE EXTENDED customers;
```

## Insert Data

```python
%sql
INSERT INTO customers VALUES
  (1, 'Ravi', 'Hyderabad'),
  (2, 'Anita', 'Bangalore');
```

## Confirm the data

```python
%sql
SELECT * FROM customers;
```

## Check version history

```python
%sql
DESCRIBE HISTORY customers
```

##  The moment of truth: drop the table

```python
%sql
DROP TABLE customers;
```

## Try to query it 

```python
%sql
SELECT * FROM customers;
```

##  The safety net: bring it back

```python
%sql
UNDROP TABLE customers;
```

##  Confirm it's back, data intact

```python
%sql
SELECT * FROM customers;
```
